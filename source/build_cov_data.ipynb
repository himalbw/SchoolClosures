```python
import pandas as pd
from pathlib import Path
```

```python
df = pd.read_csv("/Users/abhayduggirala/Desktop/harvard_senior/random/econometrics_game/SchoolClosures/Data/raw/controls/seda_cov_admindist_annual_2024.3.csv")
```

```python
df.columns
```

```python
keep = ["year", "sedaadmin", 'totenrl', 'perblk', 'lninc50all', 'povertyall', 'unempall', 'urban', 'town', 'suburb', 'rural']

df = df[keep]

df.rename(columns={"totenrl": "enrollment_dist", "perblk": "perc_black", 'lninc50all': "log_median_income", 'povertyall': "poverty_rate", 'unempall': "unemployment_rate"}, inplace=True)
```

```python
df = df[df["year"].between(2015, 2019)]
```

```python
df_collapsed = df.groupby("sedaadmin").agg({
    "enrollment_dist": "mean",
    "perc_black": "mean",
    "log_median_income": "mean",
    "poverty_rate": "mean",
    "unemployment_rate": "mean",
    "urban": "last",
    "town": "last",
    "suburb": "last",
    "rural": "last"
}).reset_index()
```

```python
df_collapsed.to_csv("/Users/abhayduggirala/Desktop/harvard_senior/random/econometrics_game/SchoolClosures/Data/clean/admin_district_controls.csv", index=False)
```
